```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer = 'word', ngram_range = (1, 2), min_df = 0, stop_words = 'english')
tfidf_matrix = tf.fit_transform(df['content'])
```

```python
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
print(cosine_similarities)
```

```python
def removing_space(text):
    text = text.strip()
    text = text.split()
    return " ".join(text)

ps = PorterStemmer()

df['content'] = df['content'].apply(lambda x: re.sub('[^a-zA-Z][^éèçàêîûô]\'', ' ', x))
df['content'] = df['content'].apply(removing_space)
df['content'] = df['content'].apply(lambda x: x.lower())
df['content'] = df['content'].apply(lambda x: x.split())
df['content'] = df['content'].apply(lambda x: [word for word in x if not word in set(stopwords.words('french'))])
```

```python
df.reset_index(drop=True)
df.head()
```

```python
results = {}
for idx, row in df.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-10:-1]
    similar_items = [(cosine_similarities[idx][i], df['id'][i]) for i in similar_indices]
    results[row['id']] = similar_items[1:]
```

```python
def item(id):
    name   = df.loc[df['id'] == id]['content'].tolist()[0].split(' // ')[0]
    desc   = ' \nDescription: ' + df.loc[df['id'] == id]['content'].tolist()[0].split(' // ')[1][0:165] + '...'
    prediction = name  + desc
    return prediction

def recommend(item_id, num):
    print('Recommending ' + str(num) + ' products similar to ' + item(item_id))
    print('---')
    recs = results[item_id][:num]
    for rec in recs:
        print('\nRecommended: ' + item(rec[1]) + '\n(score:' + str(rec[0]) + ')')
```

```python
recommend(item_id = 258580, num = 5)
```
